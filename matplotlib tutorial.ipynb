```python
crop_prod = pd.read_csv('C:/Users/bhave/Downloads/DP_LIVE_30082019023144522.csv')
crop_prod.head()
```

```python
crop_prod['TIME'] = pd.Categorical(crop_prod['TIME'])
years = crop_prod[crop_prod['SUBJECT']== 'RICE']['TIME']
rice = crop_prod[crop_prod['SUBJECT']=='RICE']['Value']
wheat = crop_prod[crop_prod['SUBJECT']== 'WHEAT']['Value']
maize = crop_prod[crop_prod['SUBJECT'] == 'MAIZE']['Value']
soybean = crop_prod[crop_prod['SUBJECT'] == 'SOYBEAN']['Value']


plt.plot(years, rice, label = 'Rice')
plt.plot(years, wheat[:4150], label = 'Wheat')
plt.plot(years, maize[:4150], label = 'Maize')
plt.plot(years, soybean[:4150], label = 'Soybean')

plt.xlabel('Year', size = 12, fontweight = 'semibold')
plt.ylabel('Thousand tonnes', size = 12, fontweight = 'semibold')
plt.title('Total OECD crop production in 1995-2016', size = 14, fontweight = 'bold')
plt.xlim(1995, 2016)
plt.grid(True, linewidth = 0.5)
plt.legend()
plt.show()
```

```python
labels_med = ['Medicine(Before)', 'Medicine(After)']
labels_placebo = ['Placebo(before)', 'Placebo(After)']

medicine = [2.88,1.42]
placebo = [2.72,2.68]
yerr_med = [0.12,0.08]
yerr_placebo = [0.24,0.13]

med_df = pd.DataFrame([medicine, placebo])
med_df.columns = ['Before', 'After']
med_df.index = ['Medicine', 'Placebo']

med_df.plot(kind = 'bar', width = 0.4, color=['midnightblue','cornflowerblue'], \
           yerr = [yerr_med, yerr_placebo])

plt.title('Effect of Medicines')
plt.xlabel('Condition')
plt.ylabel('[hsCRP] (mg/L)')
plt.xticks(rotation=0) # to keep the xtick labels horizontal
plt.legend(loc=(0.4,0.8))

plt.show()
```
